# Etapa 1 – Coleta de Dados do Diário Oficial de Avaré

## Objetivo

Esta etapa tem como objetivo realizar a coleta automatizada das publicações do Diário Oficial do Município de Avaré por meio do portal oficial da Imprensa Oficial Municipal.

O processo consiste em acessar a API disponibilizada pelo portal, recuperar os metadados das publicações e armazená-los em uma base estruturada para utilização nas etapas posteriores do projeto.

---

## Bibliotecas Utilizadas

- requests
- pandas
- json
- base64
- BeautifulSoup
- concurrent.futures
- tqdm

---

## Fluxo da Coleta

O processo executa as seguintes etapas:

1. Acessa a API do portal da Imprensa Oficial.
2. Recupera as informações de cada edição publicada.
3. Identifica os links dos documentos.
4. Organiza os metadados em um DataFrame.
5. Exporta os dados para um arquivo CSV.

---

## Dados Coletados

Entre os principais campos coletados estão:

- Identificador da publicação
- Data da publicação
- Ano
- Número da edição
- URL do documento
- Tipo do ato
- Secretaria responsável
- Quantidade de páginas
- Metadados da publicação

---

## Arquivo Gerado

Ao término da coleta é criado o arquivo:

```

data/diario_avare.csv

```

Esse arquivo será utilizado como entrada para a etapa de extração textual.

---

## Resultado Esperado

Ao final desta etapa espera-se possuir uma base organizada contendo todas as publicações disponíveis no portal, permitindo sua utilização nas etapas de processamento e treinamento do modelo de Inteligência Artificial.
