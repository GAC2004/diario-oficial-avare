# Etapa 4 – Desenvolvimento da Interface Web

## Objetivo

Esta etapa apresenta a construção da interface gráfica desenvolvida utilizando Streamlit para disponibilizar o modelo treinado aos usuários.

A aplicação permite consultar, pesquisar e classificar automaticamente publicações do Diário Oficial.

---

## Tecnologias Utilizadas

- Streamlit
- PyTorch
- Pandas

---

## Funcionalidades

A aplicação oferece as seguintes funcionalidades:

- Pesquisa por palavras-chave;
- Filtro por ano;
- Filtro por categoria;
- Visualização das publicações;
- Classificação automática utilizando Inteligência Artificial;
- Exibição da probabilidade prevista para cada categoria.

---

## Fluxo da Aplicação

O funcionamento da aplicação segue o seguinte processo:

1. Carregamento do modelo treinado.
2. Leitura da base textual.
3. Aplicação dos filtros selecionados.
4. Classificação dos textos.
5. Exibição dos resultados ao usuário.

---

## Interface

A interface foi desenvolvida priorizando simplicidade e facilidade de utilização, permitindo consultas rápidas às publicações do Diário Oficial.

---

## Resultado Esperado

Ao final desta etapa o usuário pode interagir com o sistema por meio do navegador, realizando pesquisas e obtendo automaticamente a classificação dos documentos utilizando o modelo de Inteligência Artificial desenvolvido no projeto.
