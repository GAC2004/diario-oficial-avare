# 01 - Análise Exploratória

## Importação e leitura

```
import pandas as pd
import matplotlib.pyplot as plt

df=pd.read_csv("../data/processed/dataset_final.csv")
df.head()
```

## Informações

```
df.info()
df.describe(include='all')
```

## Valores nulos

```
df.isna().sum()
```

## Distribuição de labels

```
df['label'].value_counts().plot(kind='bar')
plt.show()
```

## Documentos por ano

```
anos=df['data'].astype(str).str[:4]
anos.value_counts().sort_index().plot(kind='bar',figsize=(8,4))
plt.show()
```
