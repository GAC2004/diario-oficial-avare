# 03 - Engenharia de Atributos

## Leitura

```
import pandas as pd
import matplotlib.pyplot as plt

df=pd.read_csv("../data/processed/dataset_final.csv")
df.head()
```

## Tokens

```
df['tokens']=df['texto'].astype(str).str.split().str.len()
df['tokens'].describe()
```

## Boxplot

```
df.boxplot(column='tokens')
plt.show()
```

## Correlação

```
df[['label','tokens']].corr()
```
