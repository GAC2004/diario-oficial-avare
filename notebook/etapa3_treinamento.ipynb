# Etapa 3 – Treinamento da Rede Neural

## Objetivo

Nesta etapa é desenvolvido e treinado um modelo de Inteligência Artificial utilizando a biblioteca PyTorch para classificar automaticamente as publicações do Diário Oficial.

---

## Bibliotecas Utilizadas

- torch
- pandas
- numpy
- sklearn

---

## Classes Utilizadas

As publicações foram classificadas em quatro categorias:

- atos_normativos
- pessoal
- licitacoes_contratos
- contas_publicas

---

## Pipeline de Treinamento

O fluxo do treinamento consiste em:

1. Carregamento da base textual.
2. Vetorização dos textos.
3. Separação entre treino e teste.
4. Construção da Rede Neural.
5. Treinamento do modelo.
6. Avaliação das métricas.
7. Salvamento do modelo treinado.

---

## Modelo Desenvolvido

Foi utilizada uma Rede Neural Feedforward implementada em PyTorch.

O modelo recebe o vetor correspondente ao texto e retorna a categoria mais provável para cada publicação.

---

## Métricas Avaliadas

Durante o treinamento são calculadas:

- Accuracy
- Precision
- Recall
- F1-Score

Essas métricas permitem avaliar o desempenho do modelo na classificação automática dos documentos.
