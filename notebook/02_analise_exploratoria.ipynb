# 02 - Pré-processamento

## Leitura

```
import pandas as pd
import matplotlib.pyplot as plt

df=pd.read_csv("../data/processed/dataset_final.csv")
df.head()
```

## Comprimento dos textos

```
df['tam']=df['texto'].astype(str).str.len()
df['tam'].describe()
```

## Histograma

```
df['tam'].hist(bins=30)
plt.show()
```

## Top palavras

```
from collections import Counter
c=Counter(' '.join(df['texto'].astype(str)).split())
pd.DataFrame(c.most_common(30),columns=['palavra','freq'])
```
