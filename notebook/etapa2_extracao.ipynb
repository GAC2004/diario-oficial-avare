# Etapa 2 – Extração e Processamento dos Textos

## Objetivo

Esta etapa realiza a extração do conteúdo textual das publicações coletadas, permitindo transformar documentos PDF em texto estruturado para utilização em algoritmos de Processamento de Linguagem Natural (NLP).

---

## Bibliotecas Utilizadas

- pdfplumber
- pandas
- concurrent.futures
- re
- unicodedata

---

## Fluxo da Extração

O pipeline realiza automaticamente:

1. Leitura da base de documentos coletados.
2. Download dos arquivos PDF.
3. Extração do conteúdo textual.
4. Processamento paralelo utilizando ThreadPoolExecutor.
5. Aplicação do módulo de limpeza textual.
6. Armazenamento da base processada.

---

## Pré-processamento

Durante a extração são realizadas as seguintes operações:

- remoção de caracteres inválidos;
- remoção de múltiplos espaços;
- normalização Unicode;
- remoção de linhas vazias;
- padronização da acentuação;
- limpeza de ruídos provenientes da conversão dos PDFs.

---

## Arquivos Gerados

Após o processamento são produzidos:

```

data/processed/base_textual.csv

```

e

```

data/processed/amostra_rotulada.csv

```

---

## Resultado Esperado

Ao final desta etapa todos os documentos encontram-se convertidos em texto limpo, estruturado e preparado para utilização pelo modelo de classificação.
